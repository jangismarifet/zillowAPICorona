```python
import requests
import gspread
from oauth2client.service_account import ServiceAccountCredentials
import time
from bs4 import BeautifulSoup
```

```python
ZWSID = "Enter-your-Zillow-API-Key-Here"
citystatezip = 'Corona+CA'
address = "3398+Braemar+Ln"
allzpid = ["17766346","17764374","17784702","17883327","2092097213","69282540","52467816","17776098","52466361","17784701","2092411569","17762198","2091899447","2091899446","2091899448","2092575442","52467377","17763538","17783295","63303820","17777843","17784702","52470434","17884332","17777002","17885696","17882840","2099477599","52470252","79832345","49816662",]
zpid = []
for z in allzpid:
    if z not in zpid:
        zpid.append(z)
print ("There repeats zpids:")
print len(allzpid) - len(zpid)
count = "25"
#initialize arrays for..
location = []
price = []
street = []
city = []
zipcode = []
bath = []
room = []
lotsize = []
finsize = []
yearbuilt = []
rid = []
rname = []
rtype = []
rezpid = []
hisurl = []
valuech = []

for z in zpid:
    url = 'http://www.zillow.com/webservice/GetDeepComps.htm?zws-id='+ZWSID+'&zpid='+z+'&count='+count
    page = requests.get(url)
    tree = page.text
    # soup tree for all current values
    soup = BeautifulSoup(tree, 'lxml')
    prices = soup.find_all('amount')
    streets = soup.find_all('street')
    cities = soup.find_all('city')
    zips = soup.find_all('zipcode')
    soup.find_all('bathrooms')
    rooms = soup.find_all('bedrooms')
    lotsizes = soup.find_all('lotsizesqft')
    finsizes = soup.find_all('finishedsqft')
    yearbuilts = soup.find_all('yearbuilt')
    rids = soup.find_all('region')
    zpids = soup.find_all('zpid')
    comps = soup.find_all('comp')
    for c in comps:
        if (c.findChild('lotsizesqft')):
            lotsize.append(c.find('lotsizesqft').get_text())
        else:
            lotsize.append('N/A')
    for c in comps:
        if (c.findChild('finishedsqft')):
            finsize.append(c.find('finishedsqft').get_text())
        else:
            finsize.append('N/A')
    for c in comps:
        if (c.findChild('bathrooms')):
            bath.append(c.find('bathrooms').get_text())
        else:
            bath.append('N/A')
    for c in comps:
        if (c.findChild('yearbuilt')):
            yearbuilt.append(c.find('yearbuilt').get_text())
        else:
            yearbuilt.append('N/A')
    for c in comps:
        if (c.findChild("valuechange")):
            valuech.append(c.find('valuechange').get_text())
           
    #add soup tree text into arrays
    for zp in zpids:
        if (zp.parent.name == 'comp'):
            rezpid.append(zp.get_text())
    for p in prices:
        if (p.parent.parent.name == 'comp'):
            price.append(p.get_text())
    for s in streets:
        if (s.parent.parent.name == 'comp'):
            street.append(s.get_text())
    for c in cities:
        if (c.parent.parent.name == 'comp'):
            city.append(c.get_text())
    for z in zips:
        if (z.parent.parent.name == 'comp'):
            zipcode.append(z.get_text())
    for ro in rooms:
        if (ro.parent.name == 'comp'):
            room.append(ro.get_text())
    for r in rids:
        if (r.parent.parent.name == 'comp'):
            rid.append(r.get('id'))
    for n in rids:
        if (n.parent.parent.name == 'comp'):
            rname.append(r.get('name'))
    for t in rids:
        if (t.parent.parent.name == 'comp'):
            rtype.append(r.get('type'))



#use this code to uniquify all arrays


print(soup.prettify())
#print(rtype)
            
scope = ['https://spreadsheets.google.com/feeds']
credentials = ServiceAccountCredentials.from_json_keyfile_name('coronaHousingPrices-864beb1e6330.json', scope)
gc = gspread.authorize(credentials)
```
